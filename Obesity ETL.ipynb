```python
# Read in food data and combine the data
path = "fastfood_data1.csv"
fastfood_data1_df = pd.read_csv(path, encoding="utf-8")
path = "fastfood_data2.csv"
fastfood_data2_df = pd.read_csv(path, encoding="utf-8")
combined_fastfood_data_df = pd.concat([fastfood_data1_df, fastfood_data2_df], axis=0)

# Find the extra state and remove it
combined_fastfood_data_df_list = combined_fastfood_data_df['province'].unique().tolist()
combined_fastfood_data_df_list_sorted = sorted(combined_fastfood_data_df_list)
state_abbreviated_list = ['AL', 'AK', 'AZ', 'AR', 'CA', 'CO', 'CT', 'DE', 'FL', 'GA', 'HI', 'ID', 'IL', 'IN', 'IA', 'KS', 'KY', 'LA', 'ME', 'MD', 'MA', 'MI', 'MN', 'MS', 'MO', 'MT', 'NE', 'NV', 'NH', 'NJ', 'NM', 'NY', 'NC', 'ND', 'OH', 'OK', 'OR', 'PA', 'RI', 'SC', 'SD', 'TN', 'TX', 'UT', 'VT', 'VA', 'WA', 'WV', 'WI', 'WY']
for state in combined_fastfood_data_df_list_sorted:
    if state not in state_abbreviated_list:
        print(state)
combined_fastfood_data_df_no_DC = combined_fastfood_data_df.loc[(combined_fastfood_data_df['province'] != 'DC')]

# Check to make sure it is removed
combined_fastfood_data_df_no_DC_check = sorted(combined_fastfood_data_df_no_DC['province'].unique().tolist())
len(combined_fastfood_data_df_no_DC_check)
combined_fastfood_data_df_no_DC

# Create data frame that has state abbreviations and 
state_abbreviated_list_dict = {'States' : ['AL', 'AK', 'AZ', 'AR', 'CA', 'CO', 'CT', 'DE', 'FL', 'GA', 'HI', 'ID', 'IL', 'IN', 'IA', 'KS', 'KY', 'LA', 'ME', 'MD', 'MA', 'MI', 'MN', 'MS', 'MO', 'MT', 'NE', 'NV', 'NH', 'NJ', 'NM', 'NY', 'NC', 'ND', 'OH', 'OK', 'OR', 'PA', 'RI', 'SC', 'SD', 'TN', 'TX', 'UT', 'VT', 'VA', 'WA', 'WV', 'WI', 'WY']}
state_abbreviated_df = pd.DataFrame(state_abbreviated_list_dict)
state_abbreviated_df

# Create df with full names and concat the dataframes horizontally
states_full_name = states_obe_rnk_and_Obe_Adults_2020['States'].tolist()
states_full_name_dict = {'States Full Name' : ['Alabama',
                                                 'Alaska',
                                                 'Arizona',
                                                 'Arkansas',
                                                 'California',
                                                 'Colorado',
                                                 'Connecticut',
                                                 'Delaware',
                                                 'Florida',
                                                 'Georgia',
                                                 'Hawaii',
                                                 'Idaho',
                                                 'Illinois',
                                                 'Indiana',
                                                 'Iowa',
                                                 'Kansas',
                                                 'Kentucky',
                                                 'Louisiana',
                                                 'Maine',
                                                 'Maryland',
                                                 'Massachusetts',
                                                 'Michigan',
                                                 'Minnesota',
                                                 'Mississippi',
                                                 'Missouri',
                                                 'Montana',
                                                 'Nebraska',
                                                 'Nevada',
                                                 'New Hampshire',
                                                 'New Jersey',
                                                 'New Mexico',
                                                 'New York',
                                                 'North Carolina',
                                                 'North Dakota',
                                                 'Ohio',
                                                 'Oklahoma',
                                                 'Oregon',
                                                 'Pennsylvania',
                                                 'Rhode Island',
                                                 'South Carolina',
                                                 'South Dakota',
                                                 'Tennessee',
                                                 'Texas',
                                                 'Utah',
                                                 'Vermont',
                                                 'Virginia',
                                                 'Washington',
                                                 'West Virginia',
                                                 'Wisconsin',
                                                 'Wyoming']}
states_full_name_df = pd.DataFrame(states_full_name_dict)
states_full_name_df
state_abbreviated_df
fullnames_and_abbreviations = pd.concat([states_full_name_df, state_abbreviated_df], axis=1)
fullnames_and_abbreviations

# Add full state names to the combined_fastfood_data_df_no_DC by merging it with the df above
combined_fastfood_data_df_no_DC_full_state_names = fullnames_and_abbreviations.merge(combined_fastfood_data_df_no_DC, left_on='States', right_on='province')
combined_fastfood_data_df_no_DC_full_state_names

# Drop the States and province column
cleaned_combined_fastfood_data_df_no_DC = combined_fastfood_data_df_no_DC_full_state_names.drop(columns=['States', 'province'])
cleaned_combined_fastfood_data_df_no_DC
```
